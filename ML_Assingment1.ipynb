```python
# Q1 : Create a scatter plot between cylinder vs Co2Emission (green color)
```

```python
cylinder = data['CYLINDERS'].values
Co2Emission = data['CO2EMISSIONS'].values 
plt.scatter(cylinder, Co2Emission ,c='green') 


plt.legend(['Cylinder vs Co2 emission'])
plt.xlabel("cylinder",color = "black", weight = "bold", size = 12)
plt.ylabel("Co2Emission",color = "black", weight = "bold", size = 12) 
```

```python
# Q2 : using scatter plot compare data cylinder vs Co2Emission and Enginesize Vs Co2Emission using different colors
```

```python
cylinder = data['CYLINDERS'].values
Co2Emission = data['CO2EMISSIONS'].values

Enginesize =data['ENGINESIZE'].values
Co2Emission = data['CO2EMISSIONS'].values  


plt.scatter(cylinder ,Co2Emission ,c='red')
plt.scatter(Enginesize ,Co2Emission ,c='black') 

plt.legend(['Cylinder vs Co2 emission','Engine size vs Co2 emission'])
plt.xlabel('Cylinder/Engine size',color = "black", weight = "bold", size = 12)
plt.ylabel('Co2 emission',color = "black", weight = "bold", size = 12)
```
